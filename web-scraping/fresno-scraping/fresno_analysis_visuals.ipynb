**Fresno Data Analysis**

```python
import pylab
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('fresno-cleaned3-v2.csv')
```

```python
df.head(10)
```

```python
# YEAR ANALYSIS: PART 1.1

year_group_dict = {
}

year_group = df['Death Year']
year_range = range(1900, 2023)
for year in year_range:
    year_count = 0
    for group in year_group:
        if str(year) == str(group):
            year_count += 1
    year_group_dict[str(year)] = year_count

print("Death year divisions are as follows: ")
for key, val in year_group_dict.items():
    print(key + ": " + str(val))
print("\n")

# AGE ANALYSIS: PART 1.2
age_group = df['Age Group']
infant_count, child_count, teen_count, adult_count, elder_count, unk_count, total_count = (0, 0, 0, 0, 0, 0, 0)
for group in age_group:
    total_count += 1
    if type(group) != float: # exclude NaN cells
        if group == "Infant":
            infant_count += 1
        if group == "Child":
            child_count += 1
        if group == "Adolescent":
            teen_count += 1
        elif group == "Adult":
            adult_count += 1
        elif group == "Elder":
            elder_count += 1
    else:
        unk_count += 1

age_group_dict = {
    "Infant": infant_count,
    "Child": child_count,
    "Adolescent": teen_count,
    "Adult": adult_count,
    "Elder": elder_count,
    "Unknown": unk_count,
    "Total": total_count
}

print("Age divisions are as follows: ")
for key, val in age_group_dict.items():
    print(key + ": " + str(val))
print("\n")

# SEX ANALYSIS: PART 1.3
sex_group = df['Sex']
male_count, female_count, unk_count, total_count = (0, 0, 0, 0)
for group in sex_group:
    group = str(group)
    total_count += 1
    if group.upper() == "M":
        male_count += 1
    elif group.upper() == "F":
        female_count += 1
    else:
        unk_count += 1

sex_group_dict = {
    "Male": male_count,
    "Female": female_count,
    "Unknown": unk_count,
    "Total": total_count
}

print("Sex divisions are as follows: ")
for key, val in sex_group_dict.items():
    print(key + ": " + str(val))
print("\n")

# NO RACES ARE GIVEN: PART 1.4 NOT APPLICABLE
```

**Fresno Visualizations**

```python
# Visualizing aggregate annual burial trends for Fresno
df = pd.DataFrame.from_dict(year_group_dict, orient = 'index')
df.columns = ['Number of deaths']
df = df[30:-2]
df
```

```python
fig = plt.figure(figsize = (70, 30))
ax = fig.add_subplot()
ax.set_xlabel("Year of Death")
ax.set_ylabel("Number of Deaths per Year")
ax.set_title("Annual Deaths of Indigents in Fresno, CA Between 1930 and 2020")
ax.plot(df)
```

```python
age_df = pd.DataFrame.from_dict(age_group_dict, orient = 'index')
age_df.columns = ['Count']
age_df = age_df[:-1] # exclude Total Count from dataframe
age_df
```

```python
age_plot = age_df.plot.pie(subplots=True, figsize=(30, 10), title = "Age Composition of Indigents in Fresno, San Francisco Between 1930 and 2020", autopct='%1.1f%%')
pylab.ylabel('')
```

```python
fig = px.pie(age_df, values=[2478, 312, 102, 4143, 3212, 1471], names=['Infant', 'Child', 'Adolescent', 'Adult', 'Elder', 'Unknown'], color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_layout(title="Age Composition of Indigents in Fresno, San Francisco Between 1930 and 2020")
fig.show()
```

```python
sex_df = pd.DataFrame.from_dict(sex_group_dict, orient = 'index')
sex_df.columns = ['Count']
sex_df = sex_df[:-1]
sex_df
```

```python
sex_plot = sex_df.plot.pie(subplots=True, figsize=(30, 10), title = "Sex Composition of Indigents in Fresno, San Francisco Between 1930 and 2020", autopct='%1.1f%%')
pylab.ylabel('')
```
